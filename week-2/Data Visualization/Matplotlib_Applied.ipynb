```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from numpy.random import seed, randint
seed(100)

# Create Figure and Subplots
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=True, sharey=True, dpi=120)

# Define the colors and markers to use
colors = {0:'g', 1:'b', 2:'r', 3:'y'}
markers = {0:'o', 1:'x', 2:'*', 3:'p'}

# Plot each axes
for i, ax in enumerate(axes.ravel()):
    ax.plot(sorted(randint(0,10,10)), sorted(randint(0,10,10)), marker=markers[i], color=colors[i])  
    ax.set_title('Ax: ' + str(i))
    ax.yaxis.set_ticks_position('none')

plt.suptitle('Four Subplots in One Figure', verticalalignment='bottom', fontsize=16)    
plt.tight_layout()
# plt.show()
```

- Create a figure that has 4 sub plots on it.

- Plot 2: a scatter plot (`.scatter()`) using data `x2` and `y2` with red markers that are non-filled circles.

```python
from numpy.random import seed, randint
seed(100)


x = sorted(randint(0,10,10))
x2 = sorted(randint(0,20,10))
y = sorted(randint(0,10,10))
y2 = sorted(randint(0,20,10))

# Create Figure and Subplots
fig, axes = plt.subplots(2,2, figsize=(10,6), sharex=True, sharey=True, dpi=120)

# Define the colors and markers to use
colors = {0:'b', 1:'b', 2:'r', 3:'y'}
markers = {0:'o', 1:'x', 2:'*', 3:'p'}

axes[0][0].plot(sorted(randint(0,10,10)), sorted(randint(0,10,10)), marker='o', color='b')
axes[0][0].set_title('Graph 1')
axes[0][0].yaxis.set_ticks_position('left')

axes[0][1].scatter(sorted(randint(0,20,10)), sorted(randint(0,20,10)), marker='o', color='r')
axes[0][1].set_title('Graph 2')
axes[0][1].yaxis.set_ticks_position('left')

axes[1][0].plot(sorted(randint(0,10,10)), sorted(randint(0,10,10)), marker='o', color='b')
axes[1][0].scatter(sorted(randint(0,20,10)), sorted(randint(0,20,10)), marker='o', color='r')
axes[1][0].set_title('Graph 3')
axes[1][0].yaxis.set_ticks_position('left')

axes[1][1].plot(sorted(randint(0,10,10)), sorted(randint(0,10,10)), marker='o', color='b')
ax2 = axes[1][1].twinx()
ax2.scatter(x2, y2, marker ='o', color= 'r',facecolor= 'none')
axes[1][1].set_title('Graph 4')
axes[1][1].yaxis.set_ticks_position('both')


plt.suptitle('Four Subs in 1 Fig', verticalalignment='bottom', fontsize=20)    
plt.tight_layout()
# plt.show()
```
